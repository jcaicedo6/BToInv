```python
def plot_correlations_all(variable1, variable2, df_list=None, df_bkg=None):
    if df_bkg is not None:
        df = df_bkg
        plt.figure(figsize=(8, 6))
        sns.heatmap(df[[variable1, variable2]].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
        plt.title(f'Correlation Heatmap - Background')
        plt.show()
    elif df_list is not None:
        # Create a list of DataFrames
        dfs = df_list
        # Create a list of labels for DataFrames
        masses = [0.0, 1.0, 2.0, 3.0, 4.0, 4.5]
        labels = [f'm = {masses[i]}' for i in range(len(dfs))]

        # Create subplots based on the number of DataFrames
        fig, axes = plt.subplots(2, 3, figsize=(12, 7))
        fig.suptitle(f'Correlation Heatmaps - {variable1} vs {variable2}', fontsize=16)

        # Plot correlations for each DataFrame
        for i, (df, label) in enumerate(zip(dfs, labels)):
            row, col = divmod(i, 3)
            sns.heatmap(df[[variable1, variable2]].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1, ax=axes[row, col])
            axes[row, col].set_title(f'{label}')

        plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust subplot layout for better visualization
        plt.show()
    else:
        print("Error: Provide either df_bkg or df_list for plotting.")
# Example usage:
# Replace 'your_variable1' and 'your_variable2' with the actual column names you want to analyze
# Replace df_bkg with your background DataFrame or df_list with a list of DataFrames
# plot_correlations_all('your_variable1', 'your_variable2', df_list=[df1, df2, df3, df4, df5, df6])
# or
# plot_correlations_all('your_variable1', 'your_variable2', df_bkg=df_bkg_df)
```
