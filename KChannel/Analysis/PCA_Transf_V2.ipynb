def remove_outliers_iqr(df, whisker_width=1.5):
    Q1 = df.quantile(0.25)

    IQR = Q3 - Q1
    lower_bound = Q1 - whisker_width * IQR
    upper_bound = Q3 + whisker_width * IQR
    return df[~((df < lower_bound) | (df > upper_bound)).any(axis=1)]

```python
def PCA_Converter_rot(bkg_df, signals_df_list, var1, var2, scale=True):
    # Create the 2D array
    background_data = bkg_df[[var1, var2]].values

    # Combine signals data for rotation angle calculation
    combined_data = background_data
    for signals_df in signals_df_list:
        combined_data = np.vstack((combined_data, signals_df[[var1, var2]].values))

    if scale:
        # Standardize the data
        scaler = StandardScaler()
        data_scaled = scaler.fit_transform(combined_data)
    else:
        data_scaled = combined_data

    # Perform PCA to find principal axes and rotation angle
    pca = PCA(n_components=2)
    pca.fit(data_scaled)
    rotation_angle = np.degrees(np.arctan2(*pca.components_[0][::-1]))

    # Rotate the data points
    rotation_matrix = np.array([[np.cos(np.radians(rotation_angle)), -np.sin(np.radians(rotation_angle))],
                                [np.sin(np.radians(rotation_angle)), np.cos(np.radians(rotation_angle))]])
    rotated_data = np.dot(data_scaled, rotation_matrix)

    # Separate rotated data back into background
    background_rotated_data = rotated_data[:len(background_data)]

    # Create DataFrame with rotated data for background
    background_rotated_df = pd.DataFrame(background_rotated_data, columns=['var_x', 'var_y'])
    background_rotated_df['tag'] = bkg_df['tag'].values


    # Rotate each signal data
    signals_rotated_df_list = []
    for signals_df in signals_df_list:
        signals_data = signals_df[[var1, var2]].values
        signals_rotated_data = np.dot(signals_data, rotation_matrix)
        signals_rotated_df = pd.DataFrame(signals_rotated_data, columns=['var_x', 'var_y'])
        signals_rotated_df['tag'] = signals_df['tag'].values
        signals_rotated_df_list.append(signals_rotated_df)

    return background_rotated_df, signals_rotated_df_list
```

```python
def custom_rotation(background_df, signals_df_list, var1, var2):
    # Extract the relevant variables for background
    background_data = background_df[[var1, var2]].values

    # Combine signals data for rotation angle calculation
    combined_data = background_data
    for signals_df in signals_df_list:
        combined_data = np.vstack((combined_data, signals_df[[var1, var2]].values))

    # Calculate the means of each variable for combined data
    mean_var1 = np.mean(combined_data[:, 0])
    mean_var2 = np.mean(combined_data[:, 1])

    # Compute the covariance matrix for combined data
    cov_matrix = np.cov(combined_data, rowvar=False)

    # Eigen decomposition of the covariance matrix
    eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

    # Sort eigenvectors by eigenvalues
    sorted_indices = np.argsort(eigenvalues)[::-1]
    sorted_eigenvectors = eigenvectors[:, sorted_indices]

    # Rotate the combined data
    rotated_data = np.dot(combined_data - [mean_var1, mean_var2], sorted_eigenvectors) + [mean_var1, mean_var2]

    # Separate rotated data back into background
    background_rotated_data = rotated_data[:len(background_data)]

    # Create DataFrame with rotated data for background
    background_rotated_df = pd.DataFrame(background_rotated_data, columns=['var_x', 'var_y'])
    background_rotated_df['tag'] = background_df['tag'].values

    # Rotate each signal data
    signals_rotated_df_list = []
    for signals_df in signals_df_list:
        signals_data = signals_df[[var1, var2]].values
        signals_rotated_data = np.dot(signals_data - [mean_var1, mean_var2], sorted_eigenvectors) + [mean_var1, mean_var2]
        signals_rotated_df = pd.DataFrame(signals_rotated_data, columns=['var_x', 'var_y'])
        signals_rotated_df['tag'] = signals_df['tag'].values
        signals_rotated_df_list.append(signals_rotated_df)

    return background_rotated_df, signals_rotated_df_list
```

```python
def whitining_PCA(background_df, signals_df_list, var1, var2):
    # Extract the relevant variables for background
    background_data = background_df[[var1, var2]].values

    # Combine signals data for rotation angle calculation
    combined_data = background_data
    for signals_df in signals_df_list:
        combined_data = np.vstack((combined_data, signals_df[[var1, var2]].values))

    # Calculate the means of each variable for combined data
    mean_var1 = np.mean(combined_data[:, 0])
    mean_var2 = np.mean(combined_data[:, 1])

    # Compute the covariance matrix for combined data
    cov_matrix = np.cov(combined_data, rowvar=False)

    # Eigen decomposition of the covariance matrix
    eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

    # Sort eigenvectors by eigenvalues
    sorted_indices = np.argsort(eigenvalues)[::-1]
    sorted_eigenvectors = eigenvectors[:, sorted_indices]

    # Whitening transformation
    whitening_matrix = np.diag(1.0 / np.sqrt(eigenvalues[sorted_indices]))
    whitened_data = np.dot(combined_data - [mean_var1, mean_var2], np.dot(sorted_eigenvectors, whitening_matrix))

    # Rotate the whitened data
    rotated_data = np.dot(whitened_data, sorted_eigenvectors.T) + [mean_var1, mean_var2]

    # Separate rotated data back into background
    background_rotated_data = rotated_data[:len(background_data)]

    # Create DataFrame with rotated data for background
    background_rotated_df = pd.DataFrame(background_rotated_data, columns=['var_x', 'var_y'])
    background_rotated_df['tag'] = background_df['tag'].values

    # Rotate each signal data
    signals_rotated_df_list = []
    for signals_df in signals_df_list:
        signals_data = signals_df[[var1, var2]].values
        signals_rotated_data = np.dot(signals_data - [mean_var1, mean_var2], np.dot(sorted_eigenvectors, whitening_matrix)) + [mean_var1, mean_var2]
        signals_rotated_df = pd.DataFrame(signals_rotated_data, columns=['var_x', 'var_y'])
        signals_rotated_df['tag'] = signals_df['tag'].values
        signals_rotated_df_list.append(signals_rotated_df)

    return background_rotated_df, signals_rotated_df_list
```

```python
def Construct_Ellipse(rotated_df, var1, var2, confidence_level):
    
    # To numpy array
    rotated_data = rotated_df[[var1, var2]].to_numpy()
    
    # Calculate covariance matrix of rotated data
    covariance_rotated = np.cov(rotated_data, rowvar=False)

    # Calculate eigenvalues and eigenvectors of the covariance matrix
    eigenvalues, eigenvectors = np.linalg.eig(covariance_rotated)

    # Sort eigenvalues and eigenvectors
    sorted_indices = np.argsort(eigenvalues)[::-1]
    eigenvalues = eigenvalues[sorted_indices]
    eigenvectors = eigenvectors[:, sorted_indices]

    # Determine rotation angle and scale factors for the ellipse
    scale_factor = np.sqrt(-2 * np.log(1 - confidence_level))
    ellipse_width = scale_factor * np.sqrt(eigenvalues[0])
    ellipse_height = scale_factor * np.sqrt(eigenvalues[1])
    ellipse_angle = np.degrees(np.arctan2(*eigenvectors[:, 0][::-1]))

    center_x = np.mean(rotated_data[:, 0])
    center_y = np.mean(rotated_data[:, 1])

    # Define the equation of the ellipse
    def ellipse_equation(x, y):
        return (((x - center_x) * np.cos(np.radians(ellipse_angle)) - (y - center_y) * np.sin(np.radians(ellipse_angle))) / ellipse_width)**2 + \
               (((x - center_x) * np.sin(np.radians(ellipse_angle)) + (y - center_y) * np.cos(np.radians(ellipse_angle))) / ellipse_height)**2

    return center_x, center_y, ellipse_width, ellipse_height, ellipse_angle
```

```python
def Label_Data_Inside_Ellipse(df, var1, var2, center_x, center_y, ellipse_width, ellipse_height, ellipse_angle):
    # Define the equation of the ellipse
    def ellipse_equation(x, y):
        return (((x - center_x) * np.cos(np.radians(ellipse_angle)) - (y - center_y) * np.sin(np.radians(ellipse_angle))) / ellipse_width)**2 + \
               (((x - center_x) * np.sin(np.radians(ellipse_angle)) + (y - center_y) * np.cos(np.radians(ellipse_angle))) / ellipse_height)**2

    # Calculate the ellipse equation for each data point in the new DataFrame
    ellipse_values = ellipse_equation(df[var1], df[var2])

    # Determine the label for each data point
    inside_ellipse = ellipse_values <= 1
    inside_ellipse = inside_ellipse.astype(int)

    # Add labels to the new DataFrame
    df['Inside_Ellipse'] = inside_ellipse

    return df
```

    # Rotate the data points
    rotation_matrix = np.array([[np.cos(np.radians(rotation_angle)), -np.sin(np.radians(rotation_angle))],
                                [np.sin(np.radians(rotation_angle)), np.cos(np.radians(rotation_angle))]])
    rotated_data_b = np.dot(data_scaled_b, rotation_matrix)
    rotated_data_s = np.dot(data_scaled_s, rotation_matrix)

    sig_rot_df = pd.DataFrame(rotated_data_s, columns=['var_x', 'var_y'])
    bkg_rot_df = pd.DataFrame(rotated_data_b, columns=['var_x', 'var_y'])

    # Calculate covariance matrix of rotated data
    covariance_rotated = np.cov(rotated_data_b, rowvar=False)

    # Calculate eigenvalues and eigenvectors of the covariance matrix
    eigenvalues, eigenvectors = np.linalg.eig(covariance_rotated)

    # Sort eigenvalues and eigenvectors
    sorted_indices = np.argsort(eigenvalues)[::-1]
    eigenvalues = eigenvalues[sorted_indices]
    eigenvectors = eigenvectors[:, sorted_indices]

    # Determine rotation angle and scale factors for the ellipse
    scale_factor = np.sqrt(-2 * np.log(1 - confidence_level))
    ellipse_width = scale_factor * np.sqrt(eigenvalues[0])
    ellipse_height = scale_factor * np.sqrt(eigenvalues[1])
    ellipse_angle = np.degrees(np.arctan2(*eigenvectors[:, 0][::-1]))

    center_x = np.mean(rotated_data_b[:, 0])
    center_y = np.mean(rotated_data_b[:, 1])

    return sig_rot_df, bkg_rot_df, center_x, center_y, ellipse_width, ellipse_height, ellipse_angle

```python
import ROOT
from IPython.display import Image

def PlotInRoot(sig_df, bkg_df, var1, var2):
    # Initialize ROOT
    ROOT.PyConfig.IgnoreCommandLineOptions = True
    ROOT.gROOT.SetBatch(ROOT.kTRUE) # Suppresses the graphical display

    # Create a canvas
    canvas = ROOT.TCanvas("canvas", "2D Distribution Plot", 1000, 600)  # Increase canvas width

    # Define histogram parameters
    nbinsX = 100
    nbinsY = 100

    X_min_sig = -5
    X_max_sig = 24
    Y_min_sig = 1.2
    Y_max_sig = -0.8
    
    X_min_bkg = -5
    X_max_bkg = 24
    Y_min_bkg = -1.2
    Y_max_bkg = -0.8

    # Create histograms for signal and background
    h_sig = ROOT.TH2F("hist_signal", "Signal Distribution", nbinsX, X_min_sig, X_max_sig, nbinsY, Y_min_sig, Y_max_sig)
    h_bkg = ROOT.TH2F("hist_background", "Background Distribution", nbinsX, X_min_bkg, X_max_bkg, nbinsY, Y_min_bkg, Y_max_bkg)
    # Fill histograms with data
    sig_arr = sig_df[[var1, var2]].to_numpy()
    bkg_arr = bkg_df[[var1, var2]].to_numpy()

    for point in sig_arr:
        h_sig.Fill(point[0], point[1])

    for point in bkg_arr:
        h_bkg.Fill(point[0], point[1])

    # Set fill colors
    # h_sig.SetFillColor(ROOT.kBlue)
    # h_bkg.SetFillColor(ROOT.kRed)

    # Divide canvas into two pads
    canvas.Divide(2)  # Divide canvas into two pads
    canvas.cd(1)

    # ROOT.gStyle.SetPalette(1)
    ROOT.gStyle.SetOptStat(0)

    # Draw histograms
    # Set labels and titles
    h_bkg.GetXaxis().SetTitleOffset(1.2)  # Adjust X-axis label position
    h_bkg.GetYaxis().SetTitleOffset(1.2)  # Adjust Y-axis label position
    h_bkg.GetXaxis().SetTitle(var1)
    h_bkg.GetYaxis().SetTitle(var2)
    h_bkg.SetTitle("2D Background Distribution Plot")

    h_bkg.Draw("CONTZ")  # Use COLZ to display the color bar

    # Draw ellipse
    ellipse = ROOT.TEllipse(center_x, center_y, ellipse_width, ellipse_height)
    ellipse.SetFillColor(0)
    ellipse.SetLineColor(ROOT.kGreen)  # Set ellipse color
    ellipse.SetFillStyle(0)
    ellipse.SetLineWidth(2)
    ellipse.SetTheta(ellipse_angle)  # Set angle of rotation
    ellipse.Draw("same")

    # Draw histograms
    canvas.cd(2)

    # Set labels and titles
    h_sig.GetXaxis().SetTitleOffset(1.2)  # Adjust X-axis label position
    h_sig.GetYaxis().SetTitleOffset(1.2)  # Adjust Y-axis label position
    h_sig.GetXaxis().SetTitle(var1)
    h_sig.GetYaxis().SetTitle(var2)
    h_sig.SetTitle("2D Signal Distribution Plot")

    h_sig.Draw("CONTZ")  # Use CONTZ to overlay signal on top
    ellipse.Draw("same")


    # Update canvas
    canvas.Update()

    # Save the plot as an image file
    canvas.SaveAs("2D_Distribution_Plot.png")

    # Close the canvas to release memory
    canvas.Close()

    # Display the saved image
    return Image(filename="2D_Distribution_Plot.png")
```

```python

def plot_data_inside_ellipse_with_label(histogram, ellipse_params):
    center_x, center_y, ellipse_width, ellipse_height, ellipse_angle = ellipse_params
    
    # Create a canvas
    c2 = ROOT.TCanvas("c2", "c2", 2000, 800)
    c2.Divide(4)
    
    # Draw each histogram
    histograms = [hS_m00_2D_pdf, hB_2D_pdf, hS_Dstar_gamma_m00_2D_pdf, hB_Dstar_gamma_2D_pdf]
    
    for i, hist in enumerate(histograms, start=1):
        c2.cd(i)
        hist.Draw("COLZ")
    
        # Iterate over each bin and set the content based on the label
        for bin_x in range(1, hist.GetNbinsX() + 1):
            for bin_y in range(1, hist.GetNbinsY() + 1):
                x = hist.GetXaxis().GetBinCenter(bin_x)
                y = hist.GetYaxis().GetBinCenter(bin_y)
                label = get_label(x, y, center_x, center_y, ellipse_width, ellipse_height, ellipse_angle)
                if label == 1:  # Inside the ellipse
                    hist.SetBinContent(bin_x, bin_y, hist.GetBinContent(bin_x, bin_y))
                else:  # Outside the ellipse
                    hist.SetBinContent(bin_x, bin_y, 0)
    
    c2.Draw()

        # Save the plot as an image file
    c2.SaveAs("2D_Distribution_Plot_Inside.png")

    # Close the canvas to release memory
    c2.Close()

    # Display the saved image
    return Image(filename="2D_Distribution_Plot_Inside.png")

# Define a function to check if a point is inside the ellipse
def is_point_inside_ellipse(x, y, center_x, center_y, width, height, angle):
    cos_angle = np.cos(np.radians(angle))
    sin_angle = np.sin(np.radians(angle))
    dx = x - center_x
    dy = y - center_y
    rotated_x = cos_angle * dx + sin_angle * dy
    rotated_y = -sin_angle * dx + cos_angle * dy
    return (rotated_x / width) ** 2 + (rotated_y / height) ** 2 <= 1

# Define a function to get the label for a point based on whether it's inside or outside the ellipse
def get_label(x, y, center_x, center_y, width, height, angle):
    if is_point_inside_ellipse(x, y, center_x, center_y, width, height, angle):
        return 1  # Inside the ellipse
    else:
        return 0  # Outside the ellipse

# Assuming you have the ellipse parameters already calculated
ellipse_params = [center_x, center_y, ellipse_width, ellipse_height, ellipse_angle]

# Plot the histograms with the data inside the ellipse
plot_data_inside_ellipse_with_label(hS_m00_2D_pdf, ellipse_params)
```
