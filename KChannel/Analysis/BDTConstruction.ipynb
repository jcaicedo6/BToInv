```python
def splitDataFrame(df, fraction):
    # Shuffle the DataFrame rows
    shuffled_df = df.sample(frac=1, random_state=1)
    
    # Calculate the number of rows for the training set
    train_size = int(fraction * len(shuffled_df))
    
    # Split the DataFrame into training and other parts
    train_df = shuffled_df.iloc[:train_size]
    other_df = shuffled_df.iloc[train_size:]
    
    return train_df, other_df
```

```python
def getSample(df, fraction):
    data = df.sample(frac=fraction, random_state=1)
    return data
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold

# Function to evaluate the models
def evaluate_model(model, X, y):
	# define the evaluation procedure
	cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
	# evaluate the model and collect the results
	scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
	return scores
```

```python
# get a list of models to evaluate
def get_models_xgb_learning_rate():
	models = dict()
	# explore learning rates from 0.1 to 2 in 0.1 increments
	for i in np.arange(0.0001, 0.1, 0.005):
		key = '%.3f' % i
		models[key] = xgb.XGBClassifier(learning_rate=i, tree_method="hist")
	return models
```

```python
models = get_models_xgb_learning_rate()
# evaluate the models and store results
results, names = list(), list()
for name, model in models.items():
	# evaluate the model
	scores = evaluate_model(model, X_train, y_train)
	# store the results
	results.append(scores)
	names.append(name)
	# summarize the performance along the way
	print('>%s %.4f (+/-%.5f)' % (name, mean(scores), std(scores)))
# plot model performance for comparison
pyplot.boxplot(results, labels=names, showmeans=True)
pyplot.show()
```

```python
# get a list of models to evaluate
def get_models_xgb_max_depth():
	models = dict()
	# explore learning rates from 0.1 to 2 in 0.1 increments
	for i in np.arange(1, 21, 1):
		key = '%.3f' % i
		models[key] = xgb.XGBClassifier(max_depth=i, tree_method="hist")
	return models
```

```python
def plot_xgboost_response(clf, X_train, y_train, X_test, y_test, bins=30):
    decisions = []
    
    for X, y in [(X_train, y_train), (X_test, y_test)]:
        d1 = clf.predict_proba(X[y > 0.5])[:, 1].ravel()
        d2 = clf.predict_proba(X[y < 0.5])[:, 1].ravel()
        decisions += [d1, d2]

    low = min(np.min(d) for d in decisions)
    high = max(np.max(d) for d in decisions)
    low_high = (low, high)

    plt.hist(decisions[0],
             color='r', alpha=0.5, range=low_high, bins=bins,
             histtype='stepfilled', density=True,
             label='S (train)')
    plt.hist(decisions[1],
             color='b', alpha=0.5, range=low_high, bins=bins,
             histtype='stepfilled', density=True,
             label='B (train)')

    hist, bins = np.histogram(decisions[2],
                              bins=bins, range=low_high, density=True)
    scale = len(decisions[2]) / sum(hist)
    err = np.sqrt(hist * scale) / scale

    width = (bins[1] - bins[0])
    center = (bins[:-1] + bins[1:]) / 2
    plt.errorbar(center, hist, yerr=err, fmt='o', c='r', label='S (test)')

    hist, bins = np.histogram(decisions[3],
                              bins=bins, range=low_high, density=True)
    scale = len(decisions[2]) / sum(hist)
    err = np.sqrt(hist * scale) / scale

    plt.errorbar(center, hist, yerr=err, fmt='o', c='b', label='B (test)')

    plt.xlabel("XGBoost output")
    plt.ylabel("Arbitrary units")
    plt.legend(loc='best')
    plt.show()

# Example usage:
# Replace X_train, y_train, X_test, y_test with your actual training and testing data
# clf is your trained XGBoost classifier
# plot_xgboost_response(clf, X_train, y_train, X_test, y_test)
```

```python
def plot_BDT_cut(x_at_max, ngenSig, bdtS, bdtB, dataS, dataB, Variable):
    # Extract the variable indices for 'Mmin', 'Mmax', 'Xps', and 'q2'
    #var_indices = {'Mmin': 0, 'Mmax': 1, 'Xps': 2, 'q2': 3}
    #var_indices = {'Xps': 0, 'q2': 1}
    var_indices = {'Mmin2': 0, 'Mmax2': 1}
    var_index = var_indices[Variable]

    # Filter data based on BDT cut and variable
    VarSig = [dataS[i][var_index] for i in range(len(bdtS)) if bdtS[i] > x_at_max]
    VarBkg = [dataB[i][var_index] for i in range(len(bdtB)) if bdtB[i] > x_at_max]

    finalSig = len(VarSig)
    finalBkg = len(VarBkg)

    print(f"Signal efficiency epsilon: {100 * finalSig / ngenSig:.2f}%")

    VarSig_w = np.full_like(VarSig, 1 / finalSig) if finalSig != 0 else []
    VarBkg_w = np.full_like(VarBkg, 1 / finalSig) if finalSig != 0 else []

    plt.figure(figsize=(10, 8))

    # Plot histograms
    plt.hist([VarSig, VarBkg], weights=[VarSig_w, VarBkg_w], bins=80, range=get_hist_range(Variable),
             label=['Signal', 'Background'], hatch='//', histtype='step', lw=2)

    # Set x-axis label based on the variable
    plt.xlabel(get_x_label(Variable), fontsize=27)

    plt.ylabel("N. Events/ (0.032 GeV/$\mathrm{c}^2$)", fontsize=27)
    plt.legend(loc='upper right')
    plt.grid()
    plt.show()

def get_hist_range(variable):
    if variable == 'Mmin2':
        return (-5.0, 25.0)
    elif variable == 'Mmax2':
        return (-1.0, 24.0)
    elif variable == 'Xps':
        return (0.15, 1.2)
    elif variable == 'q2':
        return (-4.0, 24.0)

def get_x_label(variable):
    if variable == 'Mmin':
        return r"$M_{min}^2$ [GeV/$\mathrm{c}^2$]"
    elif variable == 'Mmax':
        return r"$M_{max}^2$ [GeV/$\mathrm{c}^2$]"
    elif variable == 'Xps':
        return r"$X_{ps}$"
    elif variable == 'q2':
        return r"$q^2_{rec}$"

# Example usage
# plot_BDT_cut(x_at_max, ngenSig, bdtS, bdtB, dataS, dataB, 'Mmin')
```
