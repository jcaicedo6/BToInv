```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Expected ULs Using b-only model
data = {'Luminosity': [10, 50, 100, 500, 1000, 5000, 10000, 50000],
        'AnFunc_UL': [0.00259279, 0.00118817, 0.000835967, 0.000372302, 0.00026382, 
                      0.000117692, 8.33468E-05, 3.72233E-05],
        'Argus_UL': [0.00309661, 0.00137663, 0.00097056, 0.000438689, 0.000308597,
                     0.000137951, 9.87483E-05, 4.3673E-05]}

data1 = {'Luminosity': [10, 50, 100, 500, 1000, 5000, 10000, 50000],
        'AnFunc_UL': [0.00391794, 0.00192985, 0.00085916, 0.000488214, 0.000322402, 
                      0.000118221, 6.90004E-05, 3.5853E-05],
        'Argus_UL': [0.00301327, 0.00113367, 0.00088599, 0.000423871, 0.000295078,
                     0.00015387, 8.53665E-05, 4.98511E-05]}
```

```python
UL_df = pd.DataFrame(data)
UL_df
```

```python
UL1_df = pd.DataFrame(data1)
UL1_df
```

```python
plt.rcParams.update({"font.size": 16, "font.family": "serif"})
plt.figure(figsize=(8, 7))

plt.loglog(UL_df.Luminosity, UL_df.Argus_UL, marker = 'o', c='b', ls='--', lw=1, ms=8, mew=1, mec='navy')
plt.loglog(UL_df.Luminosity, UL_df.AnFunc_UL, marker = 'v', lw=1, ms=9, mew=1, mec='g')
plt.grid(True, which="both", ls='--')

plt.xlabel('Integrated luminosity (fb'+r'$^{-1})$')
plt.ylabel('95% C.L upper limit')

plt.ylim(10E-6, 10E-3)

#plt.title("Crop Yields in Kanto")
plt.legend(['Template model', 'Parametric model']);
plt.savefig('UpperLimitsLumi.png', dpi=300)

plt.show()
```

```python
plt.rcParams.update({"font.size": 16, "font.family": "serif"})
plt.figure(figsize=(8, 7))

plt.loglog(UL1_df.Luminosity, UL1_df.Argus_UL, marker = 'o', c='b', ls='--', lw=1, ms=8, mew=1, mec='navy')
plt.loglog(UL1_df.Luminosity, UL1_df.AnFunc_UL, marker = 'v', lw=1, ms=9, mew=1, mec='g')
plt.grid(True, which="both", ls='--')

plt.xlabel('Integrated luminosity (fb'+r'$^{-1})$')
plt.ylabel('95% C.L upper limit')

plt.ylim(10E-6, 10E-3)

#plt.title("Crop Yields in Kanto")
plt.legend(['Template model', 'Analytical model']);
#plt.savefig('UpperLimitsLumi.png', dpi=300)

plt.show()
```
